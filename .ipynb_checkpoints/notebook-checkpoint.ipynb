# Implementation of data cleaning and analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import dataset
pet_supplies = pd.read_csv('pet_supplies_2212.csv')
# Sneak peak into dataset
print(pet_supplies.columns)
print(pet_supplies.head(10))
```

## Checks for and implementation of data cleaning
For each column:
-  A. Check the values match the description
-  B. Find the number of missing values
-  C. Transform data to match description (if needed)

```python
## Check product_id column
### Check we have required number of unique identifiers. The number of unique product_ids should match the number of rows.
number_of_rows = pet_supplies['product_id']
unique_product_ids = pet_supplies['product_id'].unique()
print(len(number_of_rows))
print(len(unique_product_ids))
### Check we have no missing nor duplicate values
print(pet_supplies['product_id'].isnull().any()) 
print(pet_supplies['product_id'].duplicated().any())
```

```python
## Check category column
### Get all unique values and check that they match the provided values (Housing, Food, Toys, Equipment, Medicine, Accessory).
print(pet_supplies['category'].unique())
### Check there are no null values
print(pet_supplies['category'].isnull().any()) 
### Find the number of missing values by counting all rows with the value "-"
number_of_missing_values = pet_supplies[pet_supplies['category'] == '-']
print(len(number_of_missing_values['category']))
### Replace all missing values ("_") with "Unknown"
pet_supplies['category'].replace('-', 'Unknown', inplace=True)
```

```python
## Check animal column
### Get all unique values and check that they match the provided values (Dog, Cat, Fish, Bird).
print(pet_supplies['animal'].unique())
### Find the number of missing values (if any)
print(pet_supplies['animal'].isnull().any())
```

```python
## Check size column
### Get all unique values and check that they match the provided values (Small, Medium, Large).
print(pet_supplies['size'].unique())
### Find the number of missing values (if any)
print(pet_supplies['size'].isnull().any())
### Alter values to all use title case
pet_supplies.replace({'size' : { 'large' : 'Large', 'MEDIUM' : 'Medium', 'medium' : 'Medium', 'small' : 'Small', 'SMALL' : 'Small', 'LARGE' : 'Large'}}, inplace=True)
```

```python
## Check price column

### Check the data type
print(pet_supplies['price'].dtypes)
### Fix data type to be float64
pet_supplies['price'] = pd.to_numeric(pet_supplies['price'],errors='coerce')
### Check they are rounded to 2 decimal places
print(pet_supplies['price'].dtypes)

### Find the number of missing values (if any)
print(pet_supplies['price'].isnull().any())
pet_supplies['price'].value_counts(dropna=False)

### Calculate median price
median_price = pet_supplies['price'].median().round(2)
print(median_price)
### Replace all missing values with the median price
pet_supplies['price'].fillna(median_price, inplace=True)
```

```python
## Check sales column
### Check the data type
print(pet_supplies['sales'].dtypes)

### Find the number of missing values (if any)
print(pet_supplies['sales'].isnull().any())
```

```python
## Check rating column
### Get all unique values and check that they match the provided values (1 - 10).
print(pet_supplies['rating'].unique())
print(pet_supplies['rating'].dtypes)
### Find the number of missing values (if any)
print(pet_supplies['rating'].isnull().any())
pet_supplies['rating'].value_counts(dropna=False)
### Replace all missing values with 0
pet_supplies['rating'].fillna(0, inplace=True)
```

```python
## Check repeat_purchase column
### Get all unique values and check that they match the provided values (1 or 0).
print(pet_supplies['repeat_purchase'].unique())
### Find the number of missing values (if any)
print(pet_supplies['repeat_purchase'].isnull().any())
```

## Analysis of cleaned data

```python
repeat_purchases_count = pet_supplies[pet_supplies['repeat_purchase'] == 1].groupby('repeat_purchase').value_counts().sum()
non_repeat_purchases_count = pet_supplies[pet_supplies['repeat_purchase'] == 0].groupby('repeat_purchase').value_counts().sum()

height = [repeat_purchases_count, non_repeat_purchases_count]
bars = ["Repeat purchases", "Non-repeat purchases"]


plt.bar(bars, height)
plt.ylabel('Frequency')
plt.show()
```

The frequency of Repeat purchases is approximately 900, while the frequency of Non-repeat purchases is approximately 600. This clearly shows that repeat purchase items have a higher sales rate then non-repeat purchase items.

```python
## Plotting product_id on the x_axis and category on the y_axis, and using the hue paramateter to distinguish the presence of repeat purchases, from the absence of repeat purchases
fig, axes = plt.subplots()

pet_supplies['number_of_sales'] = (pet_supplies['sales'] / pet_supplies['price'])


sns.barplot(x='category', y='number_of_sales', hue='repeat_purchase', data=pet_supplies)

## Altering the legend from '0' and '1', to 'No' and 'Yes' to make it clear which are repeat purchases and which aren't
plt.legend(title='Products repeatedly purchased',  labels=['No', 'Yes'])

## Setting the color for the legend so that it corresponds correctly to the the colors shown in the bars of the chart
ax = plt.gca()
leg = ax.get_legend()
leg.legendHandles[0].set_color('blue')
leg.legendHandles[1].set_color('orange')

#plt.yticks([0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000])


axes.set_xlabel('Categories')
axes.set_ylabel('Number of Sales')
plt.xticks(rotation=90)

plt.show()
```

```python
fig, axes = plt.subplots()

ax1 = sns.barplot(x='category', y='sales', data=pet_supplies)
plt.xticks(rotation=90)
ax2 = ax1.twinx()
sns.lineplot(x='category', y='number_of_sales', data=pet_supplies, ax=ax2, color='y', ci=None, marker='o', label='Number of Product Sales')
axes.set_xlabel('Categories')
axes.set_ylabel('Total Value of Product Sales')
ax2.set_ylabel('Number of Product Sales')

plt.show()
```
